# 七桥问题的项目

```python
%%manim -v WARNING -qh bridgedot
class bridgedot(Scene):
    def construct(self):

        #---------------------------------------------------------------------------------------
        plane = NumberPlane()
        dotA = Dot(np.array([-4,3,0]),color = WHITE,radius=0.1)
        dotB = Dot(np.array([-4,0,0]),color = WHITE,radius=0.1)
        dotC = Dot(np.array([-4,-3,0]),color = WHITE,radius=0.1)
        dotD = Dot(np.array([4,0,0]),color = WHITE,radius=0.1)
        dots = VGroup(dotA,dotB,dotC,dotD)

        A = Text("A",font_size=30,color=WHITE).next_to(dotA,LEFT)
        B = Text("B",font_size=30,color=WHITE).next_to(dotB,LEFT)
        C = Text("C",font_size=30,color=WHITE).next_to(dotC,LEFT)
        D = Text("D",font_size=30,color=WHITE).next_to(dotD,RIGHT)
        words = VGroup(A,B,C,D)
        
        #---------------------------------------------------------------------------------------
        #直线
        bridgeAB = Line(np.array([-4,3,0]),np.array([-4,0,0]),stroke_width=5,color=BLUE)
        bridgeBC = Line(np.array([-4,0,0]),np.array([-4,-3,0]),stroke_width=5,color=BLUE)

        bridgeAD = Line(np.array([-4,3,0]),np.array([4,0,0]),stroke_width=5,color=BLUE)
        bridgeBD = Line(np.array([-4,0,0]),np.array([4,0,0]),stroke_width=5,color=BLUE)
        bridgeCD = Line(np.array([-4,-3,0]),np.array([4,0,0]),stroke_width=5,color=BLUE)

        straightbridges = VGroup(bridgeAB,bridgeBC,bridgeAD,bridgeBD,bridgeCD)

        #---------------------------------------------------------------------------------------
        #曲线
        bridge_AB = Line(np.array([-4,3,0]),np.array([-4,0,0]),stroke_width=5,color=BLUE,path_arc=-2)#参数buff是用来控制线与点之间的缓冲区大小的
        bridge_BC = Line(np.array([-4,0,0]),np.array([-4,-3,0]),stroke_width=5,color=BLUE,path_arc=-2)

        curvedbridges = VGroup(bridge_AB,bridge_BC)

        #---------------------------------------------------------------------------------------
        #走的路
        ab = Arrow(np.array([-4,3,0]),np.array([-4,0,0]),stroke_width=5,color=RED)
        bd = Arrow(np.array([-4,0,0]),np.array([4,0,0]),stroke_width=5,color=RED)
        ba = Arrow(np.array([-4,0,0]),np.array([-4,3,0]),stroke_width=5,color=RED,path_arc=2)
        ad = Arrow(np.array([-4,3,0]),np.array([4,0,0]),stroke_width=5,color=RED)
        da = Arrow(np.array([4,0,0]),np.array([-4,3,0]),stroke_width=5,color=RED)
        bc = Arrow(np.array([-4,0,0]),np.array([-4,-3,0]),stroke_width=5,color=RED)
        cd = Arrow(np.array([-4,-3,0]),np.array([4,0,0]),stroke_width=5,color=RED)

        #蛞蝓猫
        #cat = ImageMobject("Slugwalk.gif").scale(3)

        #---------------------------------------------------------------------------------------
        #整个图
        
        self.play(FadeIn(dots),FadeIn(words))
        self.wait()
        self.play(Create(straightbridges),run_time=4)
        self.play(Create(curvedbridges),run_time=4)
        #self.play(FadeIn(cat))
        self.bring_to_front(dots)
        self.wait()

        #---------------------------------------------------------------------------------------
        #文字说明
        sub1 = Text("1:无论一个顶点有多少条边，当我们行走的路线在通过一个顶点时，都会使用它的两条边。",font_size=26).to_edge(DOWN)
        sub2 = Text("一条在到达顶点之前使用，一条在离开顶点时使用。",font_size=26).to_edge(DOWN)
        sub1.shift(0.5*UP)
        sub12 = VGroup(sub1,sub2)
        sub3 = Text("例如路径A——B——D,在通过B点时就会使用到AB,BD两条路径。",font_size=26).to_edge(UP)

        sub4 = Text("2:当一个顶点有奇数条边时，则它必须是线路中的起点或终点。",font_size=26).to_edge(DOWN)
        sub5 = Text("例如路径A——B——A——D,定点A有三条边,他是线路的起点。",font_size=26).to_edge(DOWN)
        sub4.shift(0.5*UP)
        sub45 = VGroup(sub4,sub5)

        sub6 = Text("例如路径A——B——D——A——B——C——D",font_size=26).to_edge(DOWN)
        sub7 = Text("点A有三条边,他是线路的起点。点D也有三条边,他是线路的终点",font_size=26).to_edge(DOWN)
        sub67 = VGroup(sub6,sub7)
        sub6.shift(0.5*UP)
        
        
        #---------------------------------------------------------------------------------------
        #无论一个顶点有多少条边，当我们行走的路线在通过一个顶点时，都会使用它的两条边，一条在到达顶点之前使用，一条在离开顶点时使用。
        self.play(Write(sub12),run_time=5)
        self.wait(2)
        #---------------------------------------------------------------------------------------
        #例如路径A——B——D,在通过B点时就会使用到AB，BD两条路径。
        self.play(Write(sub3),run_time=4)
        self.wait()
        self.play(Flash(dotB),run_time=3)
        self.wait()
        self.play(Create(ab),run_time=3)
        self.play(Create(bd),run_time=3)
        self.wait()
        #---------------------------------------------------------------------------------------
        #当一个顶点有奇数条边时，则它必须是线路中的起点或终点。
        #例如路径A——B——A——D，定点A有三条边，他是线路的起点。
        self.play(FadeOut(sub12),FadeOut(sub3),FadeOut(ab),FadeOut(bd))#擦去第一个例子
        self.play(Write(sub45),run_time=5)
        self.wait()
        self.play(Create(ab),run_time=3)
        self.play(Create(ba),run_time=3)
        self.play(Create(ad),run_time=3)
        self.play(Flash(dotA),run_time=3)
        self.wait()
        #---------------------------------------------------------------------------------------
        #例如路径A——B——D——A——B——C——D。点A有三条边，他是线路的起点。点D也有三条边，他是线路的终点。
        self.play(FadeOut(sub45),FadeOut(ab),FadeOut(ba),FadeOut(ad))
        self.play(Write(sub67),run_time=5)
        self.wait()
        self.play(Create(ab),run_time=3)
        self.play(Create(bd),run_time=3)
        self.play(Create(da),run_time=3)
        self.play(Create(ab),run_time=3)
        self.play(Create(bc),run_time=3)
        self.play(Create(cd),run_time=3)
        self.play(Flash(dotD),run_time=3)
        self.wait()

```
